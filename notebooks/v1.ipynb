### First pass

**NOTE**
I'll be using random forests, for a couple reasons:
1. It trains orders of magnitude faster than SVMs. That will extremely useful when dealing with hyperparameter search
2. It deals naturally with categorical data. In fact, SVMs rely on a the projection of the data on a different space, where a decision boundary is optimal when the distance between an instance and the boundary is high. However, the very concept of distance between categorical features is ill-defined. There are some mechanisms for calculating it (i.e. enconding smartly your features), but in this case, it is simply not worth it.

Lastly, a minor drawback of using random forests that needs mentioning: RFs are great for multiclass classification, and SVMs are indeed made for 2-way classifications. In my experience, however, they hardly beat RFs in terms of performance.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ROCAUC
```

```python
x, y = df.dropna(subset=['Status']).drop(['Status', 'Id'], 1), df.Status.dropna()
```

```python
x.shape, y.shape
```

```python
rf = RandomForestClassifier()
x_tr, x_te, y_tr, y_te = train_test_split(x, y)
```

```python
m = ROCAUC(rf)
m.fit(x_tr, y_tr)
m.score(x_te, y_te)
m.poof()
```

Close to random

```python
from sklearn.metrics import classification_report
print(classification_report(y_te, rf.predict(x_te)))
```

## Let's balance out the dataset

We will be using a technique called Synthetic Minority Oversampling TEchinque (SMOTE) [1]. It performs undersampling on the majority class while generating syntetic data from the undersampled class. The syntetic data is generated by picking the k-nearest neighbors (k set to 5) of the minority class, projects them in a metric space, and picks the point on the centroid of this new space. It can handle categorical variables by using a special distance metric named e Value Difference Metric [2]

1. https://arxiv.org/pdf/1106.1813.pdf
2. https://pdfs.semanticscholar.org/f72c/bf9f16f244f5643273fa04c25e2697fe66b9.pdf

```python
from imblearn.over_sampling import SMOTE
```

```python
rf = RandomForestClassifier()
x_resamp, y_resampled = SMOTE().fit_resample(x, y)
x_tr, x_te, y_tr, y_te = train_test_split(x_resamp, y_resampled)
```

```python
m = ROCAUC(rf)
m.fit(x_tr, y_tr)
m.score(x_te, y_te)
m.poof()
```

```python
# Welp, that's better
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te, rf.predict(x_te)))
```

### One-Hot-Encoding

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(categories='auto')
x_ohe = pandas.DataFrame(enc.fit_transform(x_resamp).todense())
x_tr, x_te, y_tr, y_te = train_test_split(x_ohe, y_resampled)
```

```python
rf = RandomForestClassifier()
m = ROCAUC(rf)
m.fit(x_tr, y_tr)
m.score(x_te, y_te)
m.poof()
```

```python
# Steady improvements!
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te, rf.predict(x_te)))
```

### Hyperparameter Search

Let's start by optimizing the integer-valued parameters. I'll still be using randomized search because it is simply just faster and we have a time constraint here. 

*although if we had more time I'd surely use grid search for fine grained search*

```python
import numpy
from sklearn.model_selection import RandomizedSearchCV

rf = RandomForestClassifier()

random_grid = {'n_estimators': [int(x) for x in numpy.linspace(start = 1000, stop = 2000, num = 10)],
               'max_features': [None, 'sqrt', 'log2'],
               'max_depth': [int(x) for x in numpy.linspace(40, 110, num = 11)] + [None],
               'min_samples_split': [3, 5, 7, 9],
               'min_samples_leaf': [1, 2, 3],
               'bootstrap': [True, False]}

rf_random = RandomizedSearchCV(estimator=rf, 
                               param_distributions=random_grid, 
                               n_iter=100, 
                               cv=3, 
                               verbose=2, 
                               random_state=42, 
                               n_jobs=8)

rf_random.fit(x_ohe, y_resampled)
```

```python
rf_random.best_params_
```

```python
r_optimized = rf.set_params(**rf_random.best_params_)
```

```python
m = ROCAUC(r_optimized)
m.fit(x_tr, y_tr)
m.score(x_te, y_te)
m.poof()
```

```python
# Moaaar improvements
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te, r_optimized.predict(x_te)))
```

### Hyperparameter Search - Continued

```python
random_grid = {
  'min_weight_fraction_leaf': [0, 0.0001, 0.0002, 0.003],
  'min_impurity_decrease': [0, 0.0001, 0.0002, 0.003],
  'n_estimators': [1000],
  'min_samples_split': [5],
  'min_samples_leaf': [1],
  'max_features': ['log2'],
  'max_depth': [40],
  'bootstrap': [False]}

rf = RandomForestClassifier()

rf_random = RandomizedSearchCV(estimator=rf, 
                               param_distributions=random_grid, 
                               n_iter=100, 
                               cv=3, 
                               verbose=2, 
                               random_state=42, 
                               n_jobs=8)

rf_random.fit(x_ohe, y_resampled)
```

```python
rf = RandomForestClassifier()
r_optimized = rf.set_params(**rf_random.best_params_)
m = ROCAUC(r_optimized)
m.fit(x_tr, y_tr)
m.score(x_te, y_te)
m.poof()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te, r_optimized.predict(x_te)))
```
